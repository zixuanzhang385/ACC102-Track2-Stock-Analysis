```python
import time
import matplotlib.pyplot as plt
import numpy as np

def print_slow(text, delay=0.03):
    for char in text:
        print(char, end='', flush=True)
        time.sleep(delay)
    print()

def get_positive_float(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value < 0:
                print("Please enter a positive number.")
                continue
            return value
        except ValueError:
            print("Invalid input, please enter a number.")

def draw_radar_chart(labels, scores, title):
    scores_closed = scores + [scores[0]]
    angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
    angles_closed = angles + [angles[0]]
    
    fig, ax = plt.subplots(figsize=(6, 6.5), subplot_kw=dict(polar=True))
    ax.plot(angles_closed, scores_closed, 'o-', linewidth=2, color='#007ACC')
    ax.fill(angles_closed, scores_closed, alpha=0.25, color='#007ACC')
    
    ax.set_xticks(angles)
    ax.set_xticklabels(labels, fontsize=11)
    
    ax.set_ylim(0, 100)
    ax.set_yticks([20, 40, 60, 80, 100])
    ax.set_yticklabels(['Poor', 'Fair', 'Good', 'Great', 'Perfect'])
    
    ax.tick_params(axis='y', labelright=False, labelleft=True, pad=5)
    ax.grid(True, linestyle='--', alpha=0.7)
    
    plt.title(title, size=14, color='#333333', y=1.08, fontweight='bold')
    plt.show()

def main():
    print_slow("========================================")
    print_slow("   ACC102 Stock Valuation Assistant v7.0")
    print_slow("   (Logic Refined & Robust)")
    print_slow("========================================")
    print()

    # --- 0. 行业选择 ---
    print_slow("Please select the industry type:")
    print("1. Financial / Bank")
    print("2. Traditional / Heavy Asset")
    print("3. Tech / Light Asset")
    
    while True:
        industry_type = input("Enter 1, 2, or 3: ")
        if industry_type in ['1', '2', '3']:
            break
        print("Invalid input, please try again.")
    
    industry_names = {'1': "Financial/Bank", '2': "Traditional", '3': "Tech/Light Asset"}
    print_slow(f"Model loaded: {industry_names[industry_type]}.\n")
    time.sleep(0.5)

    # --- 1. Safety (P/E Ratio) ---
    print_slow("--- Step 1: Safety Analysis (P/E) ---")
    pe_ratio = get_positive_float("Enter P/E: ")

    if industry_type == '1': 
        if pe_ratio <= 6: safety_score = 100; safety_reason = "Very Safe"
        elif pe_ratio <= 10: safety_score = 80; safety_reason = "Safe"
        elif pe_ratio <= 15: safety_score = 60; safety_reason = "Fair" 
        else: safety_score = 40; safety_reason = "Expensive"
    else: 
        # 包含边界
        if pe_ratio <= 15: safety_score = 100; safety_reason = "Very Safe"
        elif pe_ratio <= 25: safety_score = 80; safety_reason = "Safe"
        elif pe_ratio <= 40: safety_score = 60; safety_reason = "Fair"
        else: safety_score = 30; safety_reason = "Risky"
    
    print(f"-> Safety Score: {safety_score}/100 ({safety_reason})\n")
    time.sleep(0.5)

    # --- 2. Valuation (P/B Ratio) ---
    print_slow("--- Step 2: Valuation Analysis (P/B) ---")
    pb_ratio = get_positive_float("Enter P/B: ")

    if industry_type == '1': 
        if pb_ratio <= 0.6: valuation_score = 100; valuation_reason = "Deep Value"
        elif pb_ratio <= 1.0: valuation_score = 90; valuation_reason = "Good Value"
        elif pb_ratio <= 1.5: valuation_score = 70; valuation_reason = "Fair"
        else: valuation_score = 40; valuation_reason = "Expensive"
    elif industry_type == '2': 
        if pb_ratio <= 1.2: valuation_score = 100; valuation_reason = "Undervalued"
        elif pb_ratio <= 2.0: valuation_score = 80; valuation_reason = "Fair"
        else: valuation_score = 50; valuation_reason = "Overvalued"
    else: 
        # 优化：放宽科技股 P/B 标准，增加中间档
        if pb_ratio <= 3.0: valuation_score = 100; valuation_reason = "Undervalued"
        elif pb_ratio <= 6.0: valuation_score = 90; valuation_reason = "Attractive"
        elif pb_ratio <= 10.0: valuation_score = 70; valuation_reason = "Fair"
        else: valuation_score = 50; valuation_reason = "Expensive"

    print(f"-> Valuation Score: {valuation_score}/100 ({valuation_reason})\n")
    time.sleep(0.5)

    # --- 3. Growth (Revenue Growth) ---
    print_slow("--- Step 3: Growth Analysis ---")
    growth_rate = get_positive_float("Enter Growth Rate (%): ")

    if industry_type == '1': 
        if growth_rate >= 10: growth_score = 100; growth_reason = "High Growth"
        elif growth_rate >= 5: growth_score = 80; growth_reason = "Steady"
        else: growth_score = 60; growth_reason = "Slow"
    elif industry_type == '2': 
        if growth_rate >= 15: growth_score = 100; growth_reason = "High Growth"
        elif growth_rate >= 8: growth_score = 80; growth_reason = "Steady"
        else: growth_score = 50; growth_reason = "Slow"
    else: 
        if growth_rate >= 30: growth_score = 100; growth_reason = "Hyper Growth"
        elif growth_rate >= 20: growth_score = 90; growth_reason = "High Growth"
        elif growth_rate >= 10: growth_score = 70; growth_reason = "Moderate"
        else: growth_score = 40; growth_reason = "Slow"

    print(f"-> Growth Score: {growth_score}/100 ({growth_reason})\n")
    time.sleep(0.5)

    # --- 4. Quality (ROE) ---
    print_slow("--- Step 4: Quality Analysis (ROE) ---")
    roe = get_positive_float("Enter ROE (%): ")

    if industry_type == '1': 
        if roe >= 15: quality_score = 100; quality_reason = "Excellent"
        elif roe >= 12: quality_score = 80; quality_reason = "Good"
        elif roe >= 10: quality_score = 60; quality_reason = "Average"
        else: quality_score = 30; quality_reason = "Poor"
    elif industry_type == '2': 
        if roe >= 15: quality_score = 100; quality_reason = "Excellent"
        elif roe >= 10: quality_score = 80; quality_reason = "Good"
        elif roe >= 6: quality_score = 60; quality_reason = "Average"
        else: quality_score = 30; quality_reason = "Poor"
    else: 
        if roe >= 25: quality_score = 100; quality_reason = "Excellent"
        elif roe >= 20: quality_score = 80; quality_reason = "Good"
        elif roe >= 15: quality_score = 60; quality_reason = "Average"
        else: quality_score = 30; quality_reason = "Poor"

    print(f"-> Quality Score: {quality_score}/100 ({quality_reason})\n")
    time.sleep(0.5)

    # --- 5. Dividend Yield ---
    print_slow("--- Step 5: Dividend Analysis ---")
    dividend_yield = get_positive_float("Enter Dividend Yield (%): ")

    if industry_type == '3': 
        dividend_score = 70 
        dividend_reason = "N/A (Growth Stock)"
    else: 
        if dividend_yield > 5: dividend_score = 100; dividend_reason = "High Yield"
        elif dividend_yield > 3: dividend_score = 80; dividend_reason = "Good Yield"
        elif dividend_yield > 1.5: dividend_score = 60; dividend_reason = "Average"
        else: dividend_score = 40; dividend_reason = "Low Yield"

    print(f"-> Dividend Score: {dividend_score}/100 ({dividend_reason})\n")
    time.sleep(0.5)

    # --- Final Calculation ---
    print_slow("========================================")
    print_slow("Generating Report...")
    print_slow("========================================")
    
    total_scores = [safety_score, valuation_score, growth_score, quality_score, dividend_score]
    total_score = sum(total_scores) / 5

    print(f"\n[Total Score]: {total_score:.2f} / 100")
    
    if total_score >= 80: verdict = "Strong Buy"
    elif total_score >= 60: verdict = "Buy"
    else: verdict = "Hold"
        
    print(f"[Verdict]: {verdict}")
    print("\nDetails:")
    print(f"- Safety (P/E): {safety_reason}")
    print(f"- Valuation (P/B): {valuation_reason}")
    print(f"- Growth: {growth_reason}")
    print(f"- Quality (ROE): {quality_reason}")
    print(f"- Dividend: {dividend_reason}")

    # --- 生成雷达图 ---
    print_slow("\nGenerating Radar Chart...")
    labels = ['Safety\n(P/E)', 'Valuation\n(P/B)', 'Growth', 'Quality\n(ROE)', 'Dividend']
    draw_radar_chart(labels, total_scores, f"Stock Analysis: {verdict}")

if __name__ == "__main__":
    main()
```
